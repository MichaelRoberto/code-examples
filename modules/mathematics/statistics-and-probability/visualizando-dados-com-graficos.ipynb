# 2. Criando Gráficos com Matplotlib

```python
# Cria uma lista para representar as coordenadas "X".
x_numbers = [1, 2, 3]
```

```python
# Cria uma lista para representar as coordenadas "Y".
y_numbers = [2, 4, 6]
```

```python
# Importa plot e show do módulo pylab.
from pylab import plot, show
```

```python
# Passa como argumento para a função plot():
# - Uma lista com as coordenadas "X" = 1 ao 3
# - Uma lista com as coordenadas "Y" = 2 ao 6
plot(x_numbers, y_numbers)
show() # Exibe o plot com a função show().
```

### 2.1 Marcando pontos de Intersecção no seu Gráfico

```python
# Utilize o atributo "marker" para criar pontos de Intersecção no seu gráfico.
# NOTE: Veja que o atributo "marker" recebe argumento que pode ser: o, *, x, ...
plot(x_numbers, y_numbers, marker='o')
show() # Exibe o plot com a função show().
```

### 2.4 Comparando as tendências mensais de temperatura da cidade de Nova York

```python
# Cria uma lista para representar as temperaturas de New York no ano de 2000
nyc_temp_2000 = [31.3, 37.3, 47.2, 51.0, 63.5, 71.3, 72.3, 72.7, 66.0, 57.0, 45.3, 31.1]

# Cria uma lista para representar as temperaturas de New York no ano de 2006
nyc_temp_2006 = [40.9, 35.7, 43.1, 55.7, 63.1, 71.0, 77.9, 75.8, 66.6, 56.2, 51.9, 43.6]

# Cria uma lista para representar as temperaturas de New York no ano de 2012
nyc_temp_2012 = [37.3, 40.9, 50.9, 54.8, 65.1, 71.0, 78.8, 76.7, 68.8, 58.0, 43.9, 41.5]
```

```python
# Cria um range de 0 ao 12 com a função range().
# OBS: Por padrão a função range() começa do índice = 0.
months = range(1, 13)
```

```python
# Cria um plot para representar a temperatura de 2000.
plot(months, nyc_temp_2000)
show() # Exibe o plot criado.
```

```python
# Cria um plot para representar a temperatura de 2006.
plot(months, nyc_temp_2006)
show() # Exibe o plot criado.
```

```python
# Cria um plot para representar a temperatura de 2012.
plot(months, nyc_temp_2012)
show() # Exibe o plot criado.
```

```python
# Cria um plot/gráfico para comparar as tendências mensais de temperatura da cidade de Nova York.
plot(months, nyc_temp_2000, months, nyc_temp_2006, months, nyc_temp_2012)
show() # Exibe o plot criado.
```

### 2.5 Adicionando legenda

```python
# Cria um plot/gráfico para comparar as tendências mensais de temperatura da cidade de Nova York.
plot(months, nyc_temp_2000, months, nyc_temp_2006, months, nyc_temp_2012)

# Passa uma lista para a função legend() que vai exibir
# as legendas de cada ano respectivamente.
legend([2000, 2006, 2012])
show() # Exibe o plot criado.
```

### 2.6 Adicionando títulos e labels

```python
# Import do módulo pylab:
# - plot
# - show
# - title
# - xlabel
# - ylabel
# - legend
from pylab import plot, show, title, xlabel, ylabel, legend
```

```python
# Cria um plot/gráfico para comparar as tendências mensais de temperatura da cidade de Nova York.
plot(months, nyc_temp_2000, months, nyc_temp_2006, months, nyc_temp_2012)
legend([2000, 2006, 2012]) # Adiciona legenda para cada ano respectivamente.

# Adiciona um título para o plot/gráfico.
title('Temperatura média mensal em New York City - NYC')

xlabel('Meses') # Adiciona um label nas coordenas "X".
ylabel('Temperatura') # Adiciona um label nas coordenadas "Y".

show() # Exibe o plot criado.
```

### 2.7 Criando plots/gráficos com pyplot

```python
'''
Plot/Gráfico simples usando pyplot.
'''

# Import pyplot módule do Matplotlib.
import matplotlib.pyplot

# Função para criar gráficos.
def create_graph():
    
    x_numbers = [1, 2, 3] # Lista para as coordenadas "X".
    y_numbers = [2, 4, 6] # Lista para as coordenadas "Y".

    # Cria um plot/gráfico com pyplot.plot.
    matplotlib.pyplot.plot(x_numbers, y_numbers, marker='o')
    
    # Adiciona um título para o plot/gráfico.
    matplotlib.pyplot.title("Plot/Gráfico simples com pyplot")
    
    # Adiciona uma legenda para o plot/gráfico.
    matplotlib.pyplot.legend(["Blue"])
    
    matplotlib.pyplot.xlabel("Coordenadas - X") # Adiciona um label para as coordenadas "X".
    matplotlib.pyplot.ylabel("Coordenadas - Y") # Adiciona um label para as coordenadas "Y".
    
    matplotlib.pyplot.show() # Exibe o plot criado.

if __name__ == '__main__':
    create_graph() # Chama a função create_graph().
```

### 2.8 Criando um "alias" para o pyplot

```python
'''
Plot/Gráfico simples usando pyplot.
'''
# Importa pyplot do matplotlib como "plt"
# plt é um "alias/apelido" para pyplot.
import matplotlib.pyplot as plt

# Função para criar plots/gráficos.
def create_graph():
    
    x_numbers = [1, 2, 3] # Lista para as coordenadas "X".
    y_numbers = [2, 4, 6] # Lista para as coordenadas "Y".

    # Cria um plot/gráfico com pyplot.plot.
    plt.plot(x_numbers, y_numbers, marker='o')
    
    # Adiciona um título para o plot/gráfico.
    plt.title("Plot/Gráfico simples com pyplot - alias = plt")
    
    # Adiciona uma legenda ao plot/gráfico.
    plt.legend(["Blue"])
    
    plt.xlabel("Coordenadas - X") # Adiciona um label para as coordenadas "X".
    plt.ylabel("Coordenadas - Y") # Adiciona um label para as coordenadas "Y".
    
    plt.show() # Exibe o plot criado.

if __name__ == '__main__':
    create_graph() # Chama a função create_graph().
```
